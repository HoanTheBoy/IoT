Tiền xử lý dữ liệu:
1. Xử lý dữ liệu số:

```python
def convert_price(price_str):
    if isinstance(price_str, int):  
        return price_str  # Nếu đã là số nguyên, trả về luôn
    elif isinstance(price_str, str):
        price = int("".join(filter(str.isdigit, price_str)))
        return price
    return None  # Trả về None nếu dữ liệu không hợp lệ
```

```python
df['Price'] = df['Price'].apply(convert_price)
```

# Chuyển đổi kích thước lưu trữ


```python
def convert_storage(storage_str):
    storage_str = str(storage_str).lower()
    if 'tb' in storage_str:
        value = float(storage_str.replace('tb', '')) * 1024
    elif 'gb' in storage_str:
        value = float(storage_str.replace('gb', ''))
    elif storage_str.isdigit():
        value = float(storage_str)
    else:
        return None  # Trường hợp không hợp lệ
    return value
```

```python
df['Storage'] = df['Storage'].apply(convert_storage)
```

# Chuyển đổi kích thước màn hình

```python
def convert_screen_size(screen_size_str):
    try:
        screen_size = float(screen_size_str)
        return screen_size
    except ValueError:
        return None  # Hoặc một giá trị mặc định khác nếu cần
df['ScreenSize'] = df['ScreenSize'].apply(convert_screen_size)
```

# Chuyển đổi RAM

```python
def convert_ram(ram_str):
  """Chuyển đổi chuỗi RAM thành số."""
  try:
      # Loại bỏ 'GB' và chuyển đổi sang số nguyên
      ram = int(float(str(ram_str).replace('GB', '')))
      return ram
  except ValueError:
      return None  # Hoặc một giá trị mặc định khác nếu cần
df['RAM'] = df['RAM'].apply(convert_ram)

def convert_weight(weight_str):
    """Chuyển đổi chuỗi trọng lượng thành số."""
    try:
        # Thay thế dấu phẩy bằng dấu chấm, loại bỏ 'kg' và chuyển đổi sang float
        weight = float(str(weight_str).replace(',', '.').replace('kg', '').strip())
        return weight
    except ValueError:
        return None  # Hoặc một giá trị mặc định khác nếu cần

df['Weight'] = df['Weight'].apply(convert_weight)
```

2. Xây dựng mô hình Prophet:

Prophet được thiết kế để dự đoán chuỗi thời gian, nhưng trong trường hợp này, chúng ta có thể sử dụng nó để dự đoán giá dựa trên các đặc trưng khác.

```python
 from prophet import Prophet
```

```python
def create_prophet_features(df):
      """Tạo các đặc trưng từ Prophet."""
      model = Prophet()
      # Tạo một dataframe giả để Prophet hoạt động
      prophet_df = pd.DataFrame({
          'ds': pd.to_datetime('2024-01-01') + pd.to_timedelta(range(len(df)), unit='D'),
          'y': df['Price'].values  # Giá laptop
      })
      model.fit(prophet_df)
      future = model.make_future_dataframe(periods=0)
      forecast = model.predict(future)
      forecast = forecast.set_index(prophet_df.index)
      return forecast[['yhat', 'yhat_lower', 'yhat_upper']]  # Giá trị dự đoán, khoảng tin cậy
```

# Tạo đặc trưng Prophet cho tập huấn luyện

```python
# Tạo đặc trưng Prophet cho tập huấn luyện
prophet_features_train = create_prophet_features(df.loc[X_train.index])
X_train = pd.concat([X_train.reset_index(drop=True), prophet_features_train.reset_index(drop=True)], axis=1)

# Tạo đặc trưng Prophet cho tập kiểm tra
prophet_features_test = create_prophet_features(df.loc[X_test.index])
X_test = pd.concat([X_test.reset_index(drop=True), prophet_features_test.reset_index(drop=True)], axis=1)
```
